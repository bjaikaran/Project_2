```python
import pandas as pd

from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base

Base = declarative_base()

import pymysql
pymysql.install_as_MySQLdb()

from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.orm import Session
```

```python
stock_df = pd.read_csv("Combined_Stock_Data.csv")
stock_df
```

```python
class stocks(Base):
    __tablename__="stock_data"
    id = Column(Integer, primary_key=True)
    Date = Column(String)
    Open = Column(Float)
    High = Column(Float)
    Low = Column(Float)
    Close = Column(Float)
    Adj_Close = Column(Float)
    Volume = Column(Integer)
    Symbol = Column(String)
    Event = Column(String)
```

```python
hashtag_df = pd.read_csv("Hashtags_combined.csv")
hashtag_df
```

```python
class hashtag(Base):
    __tablename__="hashtags"
    id = Column(Integer, primary_key=True)
    hashtag = Column(String)
    count = Column(Integer)
    event = Column(String)
```

```python
engine = create_engine('sqlite:///stocks_hashtags_V1.sqlite', echo=False)
Base.metadata.create_all(engine)
session = Session(engine)
```

```python
for index, row in stock_df.iterrows():
    print(row['Symbol'])
    session.add(stocks(Date = row['Date'], Open = row['Open'], High = row['High'], Low = row['Low'], Close = row['Close'], Adj_Close = row['Adj_Close'], Volume = row['Volume'], Symbol = row['Symbol'], Event = row['Event']))

for index, row in hashtag_df.iterrows():
    print(row['Hashtag'])
    session.add(hashtag(hashtag = row['Hashtag'], count = row['Count'], event = row['Event']))
```

```python
session.commit()
```
