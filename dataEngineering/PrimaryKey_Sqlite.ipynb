```python
import pandas as pd

from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base

Base = declarative_base()

import pymysql
pymysql.install_as_MySQLdb()

from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.orm import Session
```

```python
stocks_df = pd.read_csv("iex.csv")
stocks_df.head(5)
```

```python
class stocks(Base):
    __tablename__="stock_data"
    id = Column(Integer, primary_key=True)
    date = Column(String)
    open = Column(Float)
    high = Column(Float)
    low = Column(Float)
    close = Column(Float)
    volume = Column(Integer)
    symbol = Column(String)
    source = Column(String)
    event = Column(String)
```

```python
hashtag_df = pd.read_csv("Hashtags_combined.csv")
hashtag_df
```

```python
class hashtag(Base):
    __tablename__="hashtags"
    id = Column(Integer, primary_key=True)
    hashtag = Column(String)
    count = Column(Integer)
    event = Column(String)
```

```python
engine = create_engine('sqlite:///sqlite/stocks_hashtags_V1.sqlite', echo=False)
Base.metadata.create_all(engine)
session = Session(engine)
```

```python
for index, row in stocks_df.iterrows():
    print(row['Symbol'])
    session.add(stocks(date = row['Date'], open = row['Open'], high = row['High'], low = row['Low'], close = row['Close'], volume = row['Volume'], symbol = row['Symbol'], source = row['Source'], event = row['Event']))

for index, row in hashtag_df.iterrows():
    print(row['Hashtag'])
    session.add(hashtag(hashtag = row['Hashtag'], count = row['Count'], event = row['Event']))
```

```python
session.commit()
```
